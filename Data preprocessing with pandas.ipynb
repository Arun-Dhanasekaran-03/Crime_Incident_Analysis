```python
import pandas as pd
data=pd.read_csv("C:/Users/91863/Desktop/Tasks/MY PORTFOLIO PROJECTS/Crime_Incidents_in_2024.csv")
```

```python
data.head(n=5)
```

```python
#Conversion of date and timestamp into accepted format of SQL
data["REPORT_DAT"]=pd.to_datetime(data["REPORT_DAT"])
data["START_DATE"]=pd.to_datetime(data["START_DATE"])
data["END_DATE"]=pd.to_datetime(data["END_DATE"])
data["REPORT_DAT"]=data["REPORT_DAT"].dt.strftime("%Y-%m-%d %H:%M:%S")
data["START_DATE"]=data["START_DATE"].dt.strftime("%Y-%m-%d %H:%M:%S")
data["END_DATE"]=data["END_DATE"].dt.strftime("%Y-%m-%d %H:%M:%S")
```

```python
data.head(n=5)
```

```python
#Checking for null/NaN values among columns
data.isnull().sum()
```

```python
#Checking for duplicates values among columns
data.duplicated().any()
```

```python
#Data Imputation
'''Since District column is essential in this project and too much null value can skew the results,
So Data imputation can be done by matching by matching other rows's District value with same Latitudea and Longitude value'''

required_column=["DISTRICT"]
for i in required_column:
    data[i]=data.groupby(["LATITUDE","LONGITUDE"])[i].transform(lambda x:x.replace("",pd.NA).bfill().ffill())
    
data["DISTRICT"]=data["DISTRICT"].fillna("Unknown")
```

```python
#Dropping unneccesary coulumns not required in the project 
data=data.drop(["X","Y","BID","OCTO_RECORD_ID"],axis=1)
```

```python
data.to_csv("Crime_Incidents_in_2024_cleaned.csv",header=True,index=False)
```
